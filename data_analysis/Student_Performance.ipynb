```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv('D:/project/Student_Performance.csv')
dataFrame
```

```python
dataFrame.isna().sum()
```

```python
sns.set_theme(style="dark")
```

```python
(dataFrame.groupby('Hours Studied')
.agg({'Performance Index': 'mean'})
.plot
(
    ylabel = 'Performance',
    grid=True
))
```

```python
(dataFrame.groupby('Previous Scores')
.agg({'Performance Index': 'mean'})
.plot
(
    ylabel = 'Performance',
    grid=True
))
```

```python
colors = plt.get_cmap('tab10').colors

grouped = dataFrame.groupby('Extracurricular Activities')['Performance Index'].mean()

ax = grouped.plot.bar(
    ylabel = 'Performance',
    grid=True,
    color = colors
)

labels = grouped.index.tolist()
handles = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
ax.legend(handles, labels, title='Extracurricular Activities', bbox_to_anchor=(1,1))

plt.show()
```

```python
(dataFrame.groupby('Sleep Hours')
.agg({'Performance Index': 'mean'})
.plot
(
    ylabel = 'Performance',
    grid=True
))
```

```python
(dataFrame.groupby('Sample Question Papers Practiced')
.agg({'Performance Index': 'mean'})
.plot.bar
(
    ylabel = 'Performance',
    grid=True
).legend(bbox_to_anchor=(1,1)))
```
